```python
import ultralytics
import cv2
import torch
```

```python
import kagglehub
# Video disponível no Kaggle
# Download latest version
path = kagglehub.dataset_download("amogh0810/car-videos")

print("Path to dataset files:", path)
```

```python
# Baixa o modelo
model = ultralytics.YOLO("yolo11n.pt")

# Jogao o modelo para o dispositivo de computação
device = "cuda" if torch.cuda.is_available() else "cpu"
model = model.to(device)



# processa o vídeo
results = model.track(source="/root/.cache/kagglehub/datasets/amogh0810/car-videos/versions/1", conf=0.1, iou=0.7, save=True)
```

```python
# Plotar imagem com um frame com a inferência
import matplotlib.pyplot as plt
import numpy as np

plt.show(plt.imshow(cv2.cvtColor(np.squeeze(results[0].plot()), cv2.COLOR_BGR2RGB)))
```
