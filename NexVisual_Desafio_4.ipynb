```python
import ultralytics
import matplotlib.pyplot as plt
import torch
from google.colab import drive
```

```python
drive.mount('/content/drive')

path = "/content/drive/MyDrive/nexvisual_files"

image = plt.imread(path + "/desafio_4.jpeg")
plt.imshow(image)
```

```python
# Baixa o modelo de pose da YOLO.
model = ultralytics.YOLO("yolo11n-pose.pt")

device = "cuda" if torch.cuda.is_available() else "cpu"
model = model.to(device)

# Processa a imagem.
# TODO testar outras imagens.
results = model(image)  # predict on an image
```

```python
from PIL import Image
import cv2

print(results[0])

im_array = results[0].plot()
rgb_img = cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB)

plt.imshow(rgb_img)
```
